### 1. User Login Check

```python
username = input()
password = int(input())
if username == 'admin' and password == 1234:
    print('Login Successful')
else:
    print('Invalid Credentials')
```

### 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
passed = 0
failed = 0
for i in marks:
    if i<50:
        failed += 1
    else:
        passed += 1

print("The number of students passed:", passed)
print("The number of students failed:", failed)
```

### 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for i in names:
    i = i.strip().lower()
    cleaned_names.append(i)

print(cleaned_names)
```

### 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    length = len(i)
    print("message:", i)
    print("length:", length)
    if len(i)>10:
        print("The messages is greater than size 10")
```

### 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for i in logs:
    if i == 'ERROR':
        count += 1
print("Total Error messages:", count)
```
