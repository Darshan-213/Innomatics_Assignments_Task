### 1. Smart Parking Lot Management System

```python
def parking(capacity, logs):
    current_park = 0
    peak_park = 0
    for i in logs:
        if i == "IN":
            current_park += 1
        elif i == "OUT":
            if current_park > 0:
                current_park -= 1

        if current_park > peak_park:
            peak_park = current_park
    
    print("Currently parked Vehicles:",current_park)
    if current_park > capacity:
        print("Parking Capacity is Full")
    else:
        print("Parking Available")

Parking_Capacity =  50
Vehicle_Logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking(Parking_Capacity, Vehicle_Logs)
```

### 2. Online Food Delivery Time Estimator

```python
def delivery_time(distance, traffic, weather):
    base_speed = 30
    base_time = (distance / base_speed) * 60  
    
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20
    else:
        traffic_delay = 0

    if weather == "Clear":
        weather_delay = 0
    elif weather == "Rainy":
        weather_delay = 15
    else:
        weather_delay = 0

    final_eta = base_time + traffic_delay + weather_delay
    
    print("Estimated Delivery Time:", round(final_eta), "minutes")
    
delivery_time(8, "High", "Rainy")
```

### 3. Movie Theatre Seat Occupancy

```python
def theatre(total_seats, booked_seats):
    
    booked_count = len(booked_seats)
    
    occupancy_percentage = (booked_count / total_seats) * 100
    
    print("Occupancy:", round(occupancy_percentage), "%")
    
    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
    

total_seats = 200
booked_seats = [1] * 150 

theatre(total_seats, booked_seats)
```

### 4. Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    print("Average CPU Load:", round(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)
```

### 5. Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resource_usage):
    
    overused_resources = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)

    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

###  6. Online Event Registration Capacity Controller

```python
def registration_controller(capacity, total_registrations):
    
    confirmed = 0
    waitlisted = 0
    
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

registration_controller(100, 105)
```
