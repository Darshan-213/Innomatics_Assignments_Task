### Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    print("Total Likes:", total_likes)
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

likes = [250, 300, 150, 200, 150]
analyze_engagement(likes)
```

### Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    print("Average Rainfall:", int(average))
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall = [70, 75, 68, 80, 67]
required = 70
rainfall_checker(rainfall, required)
```

### Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["rahul123", "anita_01", "rahul123", "vikram99"]
detect_duplicate_accounts(users)
```

### Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


crop_prices = [1800, 2500, 1500, 3200, 1900]
filter_premium_crops(crop_prices)
```

### Problem 9: E-Learning – Student Attendance Eligibility System

```python
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    
    for status in attendance_list:
        if status == 1:
            attended += 1
    
    percentage = (attended / total_classes) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")


attendance = [1, 1, 0, 1, 1]
check_attendance(attendance)
```

### Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status

total_bill, usage_status = electricity_bill(250)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
