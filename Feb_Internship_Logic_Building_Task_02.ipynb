### 1. Unique Words in a Sentence

```python
sentence = input()
words = sentence.split()
unique = set(words)
print("The unique words in python are:",unique)
print("The length of the unique words is:", len(unique))
```

### 2. Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "kiran": 72000
}
highest_paid = max(employees, key = employees.get)
print(highest_paid, employees[highest_paid])
```

### 3. Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
print("List:",numbers)
print("Max:",max(numbers))
print("Min:",min(numbers))
```

### 4. Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for i in prices:
    if i>1000:
        count += 1
print("The Products above 1000 are:",count)
```

### 5. Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present = 0
absent = 0
for i in attendance:
    if i=="P":
        present += 1
    elif i=="A":
        absent += 1
    else:
        print("Invalid credentials")
print("The attendence percentage:",present/len(attendance)*100)
```
